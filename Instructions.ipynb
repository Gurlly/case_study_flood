### Instructions

Install the required / necessary datasets
- 1. Administrative Boundaries
    - Go to this link https://data.humdata.org/dataset/cod-ab-phl
    - Install phl_adm_psa_namria_20231106_SHP.zip
- 2. Roads / Transport Network
    - Go to this link https://download.geofabrik.de/asia/philippines.html
    - Install philippines-latest-free.shp.zip
- 3. Population
    - Go to this link https://hub.worldpop.org/geodata/summary?id=6316
    - install dataset
- 4. Critical Facilities
    - Go to this link https://data.humdata.org/dataset/hotosm_phl_health_facilities
    - Install hotosm_phl_health_facilities_points_shp.zip
- 5. Flood Hazard
    - Go to this link https://lipad-fmc.dream.upd.edu.ph/layers/geonode%3Amanila_fh100yr_10m
    - Install layer WGS84 - zipped sharefile
